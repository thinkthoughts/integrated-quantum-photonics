# 00 — Why Integrated Quantum Photonics

**Question**

Why does scaling quantum photonics shift from adding more devices to exploiting many optical modes on integrated photonic chips?

This notebook builds an executable interpretation of the seminar roadmap:

```text
Scaling pressure
→ Frequency multiplexing
→ Quantum frequency combs
→ Kerr nonlinearity
→ Two-mode squeezing
→ Many-mode entanglement
→ Integrated photonic circuits
→ Loss characterization
→ Cluster-state architectures
```


````python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import zipfile
from IPython.display import FileLink, FileLinks, display

ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent

FIG = ROOT / 'figures'
RES = ROOT / 'results'
CSV = RES / 'csv'
JS = RES / 'json'

for p in [FIG, RES, CSV, JS]:
    p.mkdir(parents=True, exist_ok=True)

print('ROOT:', ROOT)
print('FIG:', FIG)
print('RES:', RES)
````

## Figure 1 — Scaling pressure

The first pressure is straightforward: more quantum channels often means more physical components.

Integrated photonics becomes attractive because device replication does not scale gently.

````python
steps = ['More\nchannels', 'More\ndevices', 'Packaging\ncomplexity', 'Scaling\npressure']
x = np.arange(len(steps))

fig, ax = plt.subplots(figsize=(8, 2.8))
ax.scatter(x, [0] * len(x), s=200)

for i in range(len(x) - 1):
    ax.annotate('', xy=(x[i + 1], 0), xytext=(x[i], 0), arrowprops=dict(arrowstyle='->', linewidth=1.8))

for xi, s in zip(x, steps):
    ax.text(xi, -0.15, s, ha='center', va='top')

ax.set_title('Scaling Pressure in Photonic Quantum Systems')
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-0.5, 0.5)
ax.axis('off')

fig.tight_layout()
path = FIG / '00_scaling_pressure.png'
fig.savefig(path, dpi=200)
plt.show()
print('saved:', path)
````

## Figure 2 — Device scaling versus mode scaling

Device scaling increases available channels by adding physical devices.

Mode scaling asks whether one integrated device can support many usable optical modes.

````python
devices = np.arange(1, 9)
device_channels = devices
mode_channels = np.full_like(devices, 8)

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(devices, device_channels, marker='o', label='Device scaling: channels ∝ devices')
ax.plot(devices, mode_channels, marker='s', label='Mode scaling: many channels in one device')

ax.set_title('Device Scaling versus Mode Scaling')
ax.set_xlabel('Physical devices')
ax.set_ylabel('Available channels')
ax.legend()
ax.grid(True, alpha=0.3)

fig.tight_layout()
path = FIG / '00_device_vs_mode_scaling.png'
fig.savefig(path, dpi=200)
plt.show()
print('saved:', path)
````

## Figure 3 — Seminar arc

The seminar arc is not only `many modes`.

It moves through nonlinear enhancement, squeezing, many-mode entanglement, integration, and loss.

````python
labels = ['Scaling', 'Freq.\nMux', 'Quantum\nComb', 'Kerr', 'Squeezing', 'Entangle', 'Integrate', 'Loss']
x = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(11, 2.8))
ax.scatter(x, [0] * len(x), s=150)

for i in range(len(x) - 1):
    ax.annotate('', xy=(x[i + 1], 0), xytext=(x[i], 0), arrowprops=dict(arrowstyle='->', linewidth=1.6))

for xi, label in zip(x, labels):
    ax.text(xi, -0.18, label, ha='center', va='top', fontsize=9)

ax.set_title('Seminar Arc: From Scaling Pressure to Loss-Limited Integration')
ax.set_xlim(-0.5, len(labels) - 0.5)
ax.set_ylim(-0.5, 0.5)
ax.axis('off')

fig.tight_layout()
path = FIG / '00_seminar_arc.png'
fig.savefig(path, dpi=200)
plt.show()
print('saved:', path)
````

## Figure 4 — Constraint transition

The useful post-seminar shift is this:

```text
early question: can we generate many modes?
later question: can we squeeze, route, measure, and preserve them?
```

So the bottleneck migrates from modal abundance to engineering constraints.

````python
cats = ['Generate\nModes', 'Squeeze', 'Route', 'Measure', 'Reduce\nLoss']
vals = [1, 2, 3, 4, 5]

fig, ax = plt.subplots(figsize=(7, 3.5))
ax.bar(cats, vals)
ax.set_title('Constraint Transition After Modal Abundance')
ax.set_ylabel('Relative engineering emphasis')
ax.grid(True, axis='y', alpha=0.3)

fig.tight_layout()
path = FIG / '00_constraint_transition.png'
fig.savefig(path, dpi=200)
plt.show()
print('saved:', path)
````

## Summary table

This table records the notebook's interpretation of the seminar arc.

````python
summary = pd.DataFrame({
    'Stage': ['Scaling', 'Multiplexing', 'Comb', 'Kerr', 'Squeezing', 'Entanglement', 'Integration', 'Loss'],
    'Focus': ['Need channels', 'Modes', 'Resonator', 'Interaction', 'Quadrature resources', 'Graph resources', 'Chip', 'Measurement and preservation'],
    'Question': [
        'Why does replication become expensive?',
        'Can frequency carry many channels?',
        'Can one resonator generate many modes?',
        'Can nonlinear interaction be enhanced?',
        'Can modes become quantum resources?',
        'Can pair resources become multipartite structure?',
        'Can the architecture fit on chip?',
        'Can loss be characterized and reduced?'
    ]
})

summary.to_csv(CSV / '00_summary.csv', index=False)
summary.to_json(JS / '00_summary.json', orient='records', indent=2)
summary
````

## Download outputs

This cell packages the notebook outputs and displays clickable links in Jupyter / Colab.

The important fix is that `FileLink` and `FileLinks` receive string paths.

````python
zip_path = RES / '00_outputs.zip'

files_to_zip = [
    FIG / '00_scaling_pressure.png',
    FIG / '00_device_vs_mode_scaling.png',
    FIG / '00_seminar_arc.png',
    FIG / '00_constraint_transition.png',
    CSV / '00_summary.csv',
    JS / '00_summary.json',
]

with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as z:
    for file in files_to_zip:
        if file.exists():
            z.write(file, file.relative_to(ROOT))

print(f'Created: {zip_path}')

print('\nDownload ZIP:')
display(FileLink(str(zip_path)))

print('\nBrowse results directory:')
display(FileLinks(str(RES)))

summary
````

## Takeaway

Integrated quantum photonics is not only about making more modes.
