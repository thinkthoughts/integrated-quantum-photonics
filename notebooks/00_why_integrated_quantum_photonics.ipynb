# 00 — Why Integrated Quantum Photonics

**Question**

Why does scaling quantum photonics shift from adding more devices to exploiting many optical modes on integrated photonic chips?

This notebook builds an executable interpretation of the seminar roadmap:

```
Scaling pressure
→ Frequency multiplexing
→ Quantum frequency combs
→ Kerr nonlinearity
→ Two-mode squeezing
→ Many-mode entanglement
→ Integrated photonic circuits
→ Loss characterization
→ Cluster-state architectures
```


````python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json, zipfile

ROOT=Path.cwd()
FIG=ROOT.parent/'figures' if ROOT.name=='notebooks' else ROOT/'figures'
RES=ROOT.parent/'results' if ROOT.name=='notebooks' else ROOT/'results'
CSV=RES/'csv'; JS=RES/'json'
for p in [FIG,CSV,JS]:
    p.mkdir(parents=True,exist_ok=True)
print(FIG)
````

## Figure 1 — Scaling pressure

````python
steps=['More\nchannels','More\ndevices','Packaging\ncomplexity','Scaling\npressure']
x=np.arange(len(steps))
fig,ax=plt.subplots(figsize=(8,2.8))
ax.scatter(x,[0]*len(x),s=200)
for i in range(len(x)-1):
    ax.annotate('',xy=(x[i+1],0),xytext=(x[i],0),arrowprops=dict(arrowstyle='->'))
for xi,s in zip(x,steps):
    ax.text(xi,-0.15,s,ha='center')
ax.set_xlim(-0.5,3.5); ax.set_ylim(-0.5,0.5); ax.axis('off')
fig.tight_layout()
fig.savefig(FIG/'00_scaling_pressure.png',dpi=200)
plt.show()
````

## Figure 2 — Device scaling versus mode scaling

````python
fig,ax=plt.subplots(figsize=(7,4))
ax.plot([1,8],[1,8],label='Devices')
ax.plot([1,8],[8,8],label='Modes per device')
ax.set_xlabel('Physical devices')
ax.set_ylabel('Available channels')
ax.legend()
fig.tight_layout()
fig.savefig(FIG/'00_device_vs_mode_scaling.png',dpi=200)
plt.show()
````

## Figure 3 — Seminar arc

````python
labels=['Scaling','Freq.\nMux','Quantum\nComb','Kerr','Squeezing','Entangle','Integrate','Loss']
fig,ax=plt.subplots(figsize=(11,2.6))
x=np.arange(len(labels))
ax.scatter(x,[0]*len(x),s=150)
for i in range(len(x)-1):
    ax.annotate('',xy=(x[i+1],0),xytext=(x[i],0),arrowprops=dict(arrowstyle='->'))
for xi,l in zip(x,labels):
    ax.text(xi,-0.18,l,ha='center',fontsize=9)
ax.axis('off')
fig.tight_layout()
fig.savefig(FIG/'00_seminar_arc.png',dpi=200)
plt.show()
````

## Figure 4 — Constraint transition

````python
cats=['Generate\nModes','Squeeze','Route','Measure','Reduce\nLoss']
vals=[1,2,3,4,5]
fig,ax=plt.subplots(figsize=(7,3.5))
ax.bar(cats,vals)
ax.set_ylabel('Relative engineering emphasis')
fig.tight_layout()
fig.savefig(FIG/'00_constraint_transition.png',dpi=200)
plt.show()

summary=pd.DataFrame({
'Stage':['Scaling','Multiplexing','Comb','Kerr','Entanglement','Integration','Loss'],
'Focus':['Need channels','Modes','Resonator','Interaction','Graph resources','Chip','Measurement']
})
summary.to_csv(CSV/'00_summary.csv',index=False)
summary.to_json(JS/'00_summary.json',orient='records',indent=2)

zip_path=RES/'00_outputs.zip'
with zipfile.ZipFile(zip_path,'w') as z:
    for f in FIG.glob('00_*.png'):
        z.write(f,f.name)
    z.write(CSV/'00_summary.csv','00_summary.csv')
    z.write(JS/'00_summary.json','00_summary.json')
print('Download:',zip_path)
summary
````
