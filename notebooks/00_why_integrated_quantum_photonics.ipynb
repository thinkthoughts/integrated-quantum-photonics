# 00 — Why Integrated Quantum Photonics?

**Engineering question**

Why does quantum photonics shift from adding more physical devices to exploiting many optical modes on integrated photonic chips?

This notebook is the anchor for the repo. It turns the seminar-level narrative into a compact executable roadmap:

```text
Scaling pressure
→ Frequency multiplexing
→ Quantum frequency combs
→ Kerr nonlinear interaction
→ Two-mode squeezing
→ Many-mode entanglement
→ Integrated photonic circuits
→ Loss characterization
→ Applications
```

The point is not only that one chip can contain many modes.

The point is that useful quantum scale depends on whether those modes can be squeezed, routed, measured, preserved, and connected into larger resources.

- from the repo root, or

It creates:

````python
from pathlib import Path
import json
import zipfile

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

FIG = ROOT / "figures"
RES = ROOT / "results"
CSV = RES / "csv"
JS = RES / "json"

for p in [FIG, RES, CSV, JS]:
    p.mkdir(parents=True, exist_ok=True)

def savefig(fig, filename, dpi=220):
    path = FIG / filename
    fig.tight_layout()
    fig.savefig(path, dpi=dpi)
    print(f"saved: {path}")
    return path

print("ROOT:", ROOT)
print("FIG:", FIG)
print("RES:", RES)
````

## 1. Scaling pressure

The seminar begins from a broad scaling problem:

- quantum computing wants many qubits or qumodes,
- quantum communication wants many channels,
- quantum sensing wants many correlated resources,
- but device replication creates packaging, yield, routing, and measurement pressure.

The first figure states the motivation without pretending that the physics is already solved.


````python
steps = [
    "More quantum\nchannels",
    "Device replication\ngets expensive",
    "Frequency multiplexing\nadds modes",
    "Integrated photonics\ncontains the system",
]

x = np.arange(len(steps))
fig, ax = plt.subplots(figsize=(10, 3.2))

ax.scatter(x, np.zeros_like(x), s=260, zorder=3)
for i in range(len(x) - 1):
    ax.annotate(
        "",
        xy=(x[i + 1], 0),
        xytext=(x[i], 0),
        arrowprops=dict(arrowstyle="->", linewidth=2.2),
    )

for xi, label in zip(x, steps):
    ax.text(xi, -0.18, label, ha="center", va="top", fontsize=11)

ax.set_title("Scaling Pressure: From Replication to Integrated Modes", fontsize=16)
ax.set_xlim(-0.4, len(steps) - 0.6)
ax.set_ylim(-0.55, 0.45)
ax.axis("off")

savefig(fig, "00_scaling_pressure.png")
plt.show()
````

## 2. Device scaling versus mode scaling

A device-replication picture says:

```text
more channels → more physical devices
```

A mode-scaling picture asks:

```text
can one device carry many usable modes?
```

This figure avoids treating mode scaling as a magic flat line. Instead, it compares two ways to reach more channels:

- adding more physical devices,
- increasing modal capacity per integrated device.


````python
N = np.arange(1, 17)

device_channels = N
modes_per_device = np.array([1, 2, 4, 8, 12, 16, 24, 32, 45, 60, 75, 90, 110, 130, 150, 170])
one_device_reference = modes_per_device

fig, ax = plt.subplots(figsize=(8, 4.8))

ax.plot(N, device_channels, marker="o", label="Replication: N devices → N channels")
ax.plot(N, one_device_reference, marker="s", label="Mode scaling: richer modal capacity")

ax.set_title("Device Scaling versus Mode Scaling", fontsize=16)
ax.set_xlabel("Engineering effort / design generation")
ax.set_ylabel("Available optical quantum channels")
ax.grid(True, alpha=0.3)
ax.legend()

savefig(fig, "00_device_vs_mode_scaling.png")
plt.show()
````

## 3. Research roadmap

Before the seminar, the tempting summary was:

```text
frequency multiplexing → many modes
```

After the seminar, the stronger roadmap is:

```text
many modes
→ nonlinear interaction
→ squeezing
→ many-mode entanglement
→ integrated measurement
→ loss-limited applications
```

This figure becomes the repo's high-level map.


````python
roadmap = [
    "Scaling\npressure",
    "Frequency\nmultiplexing",
    "Quantum\nfrequency combs",
    "Kerr\ninteraction",
    "Two-mode\nsqueezing",
    "Multipartite\nentanglement",
    "Integrated\nphotonics",
    "Loss\ncharacterization",
    "Applications",
]

x = np.arange(len(roadmap))
fig, ax = plt.subplots(figsize=(13, 3.2))

ax.scatter(x, np.zeros_like(x), s=180, zorder=3)
for i in range(len(x) - 1):
    ax.annotate(
        "",
        xy=(x[i + 1], 0),
        xytext=(x[i], 0),
        arrowprops=dict(arrowstyle="->", linewidth=1.9),
    )

for xi, label in zip(x, roadmap):
    ax.text(xi, -0.18, label, ha="center", va="top", fontsize=9.5)

ax.set_title("Research Roadmap: Integrated Quantum Photonics", fontsize=16)
ax.set_xlim(-0.4, len(roadmap) - 0.6)
ax.set_ylim(-0.55, 0.45)
ax.axis("off")

savefig(fig, "00_research_roadmap.png")
plt.show()
````

## 4. Constraint transition

The later slides shifted the center of gravity.

The problem is not merely:

```text
Can we generate modes?
```

It becomes:

```text
Can we produce squeezing, route modes, characterize loss, reduce loss, and scale multipartite entanglement?
```

That transition matters because it tells us what the later notebooks should test.


````python
constraint_steps = [
    "Generate\nmodes",
    "Produce\nsqueezing",
    "Route optical\nmodes",
    "Characterize\nloss",
    "Reduce\nloss",
    "Scale multipartite\nentanglement",
]

x = np.arange(len(constraint_steps))
y = np.array([0, 1, 2, 3, 4, 5])

fig, ax = plt.subplots(figsize=(10, 4.2))

ax.plot(x, y, marker="o", linewidth=2.2)
for i in range(len(x) - 1):
    ax.annotate(
        "",
        xy=(x[i + 1], y[i + 1]),
        xytext=(x[i], y[i]),
        arrowprops=dict(arrowstyle="->", linewidth=1.5),
    )

for xi, yi, label in zip(x, y, constraint_steps):
    ax.text(xi, yi + 0.18, label, ha="center", va="bottom", fontsize=10)

ax.set_title("Constraint Transition After Modal Abundance", fontsize=16)
ax.set_xlabel("Engineering progression")
ax.set_ylabel("Downstream constraint load")
ax.set_xticks([])
ax.set_yticks([])
ax.grid(True, alpha=0.25)

savefig(fig, "00_constraint_transition.png")
plt.show()
````

## 6. Compact interpretation

The seminar reframes integrated quantum photonics as a transition from **hardware replication** to **modal resource management**.

Frequency multiplexing creates modal abundance.

Kerr nonlinear interaction turns selected modes into squeezed quantum resources.

The next constraints are not abstract: they are loss, routing, measurement, integration, and graph-state construction.

That is why the repo should not stop at frequency multiplexing. It should continue toward loss characterization and cluster-state architectures.
