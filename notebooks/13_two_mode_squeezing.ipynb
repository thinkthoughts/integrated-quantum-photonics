# Notebook 13 — Two-Mode Squeezing Generates Quantum Correlations

Engineering statement: **Nonlinear interaction generates frequency pairs whose quantum fluctuations become correlated.**

```
import numpy as np
import matplotlib.pyplot as plt
```

## Figure 1 — Four-wave mixing

```
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(8,4)); ax.axis("off")
ax.text(.18,.72,r"$\omega_0$",fontsize=28); ax.text(.18,.22,r"$\omega_0$",fontsize=28)
ax.text(.46,.48,r"$\chi^{(3)}$",fontsize=24)
ax.annotate("",(.72,.48),(.56,.48),arrowprops=dict(arrowstyle="->",lw=2))
ax.text(.82,.68,r"$\omega_{-1}$",fontsize=26); ax.text(.82,.18,r"$\omega_{+1}$",fontsize=26)
ax.text(.28,-.02,r"$2\omega_0=\omega_{-1}+\omega_{+1}$",fontsize=18)
plt.show()
```

## Figure 2 — Symmetric mode pairs

```
fig,ax=plt.subplots(figsize=(8,5)); ax.set_xlim(-6,6); ax.set_ylim(.5,5.5); ax.axis("off")
for i in range(1,6):
 ax.plot([-i,i],[i,i],lw=2); ax.scatter([-i,i],[i,i],s=100)
 ax.text(-i-.25,i+.15,fr"$\omega_{{-{i}}}$"); ax.text(i+.1,i+.15,fr"$\omega_{{+{i}}}$")
ax.text(0,5.2,r"$\omega_0$",ha="center"); plt.show()
```

## Figure 3 — Nonlinear threshold

```
x=np.linspace(0,10,300); y=np.where(x<3.2,0,((x-3.2)/(10-3.2))**1.6)
plt.figure(figsize=(7,4)); plt.plot(x,y,lw=3); plt.axvline(3.2,ls="--")
plt.xlabel("Pump power"); plt.ylabel("Relative comb output"); plt.grid(alpha=.3); plt.show()
```

## Figure 4 — Scaling

```
p=np.arange(1,21)
plt.figure(figsize=(7,4)); plt.plot(p,p,marker="o")
plt.xlabel("Mode pairs"); plt.ylabel("Available squeezed pairs"); plt.grid(alpha=.3); plt.show()
```

## Summary

- Kerr nonlinearity enables four-wave mixing.
- Four-wave mixing creates symmetric frequency pairs.
- Two-mode squeezing produces quantum correlations.
- A frequency comb provides many scalable quantum channels.
